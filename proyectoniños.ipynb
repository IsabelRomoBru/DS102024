```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://saposyprincesas.elmundo.es/actividades-ninos/'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'  # A veces necesario para evitar bloqueos
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Lista para almacenar la info
eventos = []

# Cada artículo es un evento
for article in soup.find_all('article', class_='elementor-post'):
    try:
        titulo = article.find('h3').get_text(strip=True)
        enlace = article.find('a', href=True)['href']
        fecha = article.find('span', class_='meta-fecha').get_text(strip=True) if article.find('span', class_='meta-fecha') else 'No disponible'
        ubicacion = article.find('span', class_='meta-ubicacion').get_text(strip=True) if article.find('span', class_='meta-ubicacion') else 'No disponible'

        eventos.append({
            'Título': titulo,
            'Fecha': fecha,
            'Ubicación': ubicacion,
            'Enlace': enlace
        })
    except Exception as e:
        print("Error procesando evento:", e)

# Convertir a DataFrame
df = pd.DataFrame(eventos)
print(df.head())
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time

# 1. Configuración del navegador (sin headless)
options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')  # abre ventana completa

# 2. Iniciar el navegador
driver = webdriver.Chrome(options=options)
driver.get('https://saposyprincesas.elmundo.es/actividades-ninos/')

# 3. Esperar a que aparezcan los elementos de las actividades
try:
    WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.CLASS_NAME, 'elementor-post'))
    )
    print("✔ Contenido cargado")
except:
    print("❌ No se cargaron los eventos. Revisa si hay bloqueadores de cookies, captchas o banners.")
    driver.quit()

# 4. Obtener HTML de la página ya renderizada
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

# 5. Extraer actividades
eventos = []
articles = soup.find_all('article', class_='elementor-post')

for article in articles:
    h3 = article.find('h3')
    a = article.find('a', href=True)
    if h3 and a:
        eventos.append({
            'Título': h3.text.strip(),
            'Enlace': a['href']
        })

# 6. Cerrar navegador
driver.quit()

# 7. Mostrar los datos
df = pd.DataFrame(eventos)
print(df.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from IPython.display import display, HTML
import time

base_url = 'https://quehacerconlosninos.es'
main_url = base_url + '/ocio/'
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(main_url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')
eventos = soup.find_all('article')

data = []

for evento in eventos:
    titulo_tag = evento.find('h3') or evento.find('h2') or evento.find('h1')
    titulo = titulo_tag.get_text(strip=True) if titulo_tag else 'Sin título'

    link_tag = evento.find('a')
    link = link_tag['href'] if link_tag else ''
    if link and not link.startswith('http'):
        link = base_url + link

    img_tag = evento.find('img')
    imagen = img_tag['src'] if img_tag else ''

    # Variables del evento individual
    edad = ubicacion = fecha = categoria = 'No disponible'

    # Segundo nivel: scrapeo dentro del enlace
    if link != '':
        try:
            detalle = requests.get(link, headers=headers)
            detalle_soup = BeautifulSoup(detalle.content, 'html.parser')
            parrafos = detalle_soup.find_all('p')
            texto_completo = " ".join(p.get_text() for p in parrafos)

            # Búsquedas simples por texto (puedes mejorar con regex)
            if 'años' in texto_completo:
                edad = next((p.get_text() for p in parrafos if 'años' in p.get_text()), 'No disponible')
            if 'Madrid' in texto_completo or 'Calle' in texto_completo:
                ubicacion = next((p.get_text() for p in parrafos if 'Madrid' in p.get_text() or 'Calle' in p.get_text()), 'No disponible')
            if 'del' in texto_completo and 'al' in texto_completo:
                fecha = next((p.get_text() for p in parrafos if 'del' in p.get_text() and 'al' in p.get_text()), 'No disponible')

            # Categoría: buscar breadcrumb o categorías si están
            categoria_tag = detalle_soup.find('span', class_='posted_in')
            categoria = categoria_tag.get_text(strip=True) if categoria_tag else 'No disponible'

            time.sleep(1)  # Evitar sobrecarga al servidor

        except Exception as e:
            print(f"Error al procesar {link}: {e}")

    data.append({
        'Título': titulo,
        'Link': link,
        'Imagen': f'<img src="{imagen}" width="120">' if imagen else '',
        'Edad Recomendada': edad,
        'Ubicación': ubicacion,
        'Fechas': fecha,
        'Categoría': categoria
    })

# DataFrame y visualización
df = pd.DataFrame(data)
display(HTML(df.to_html(escape=False)))
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from IPython.display import display, HTML
import time

base_url = 'https://quehacerconlosninos.es'
main_url = base_url + '/ocio/'
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(main_url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')
eventos = soup.find_all('article')

data = []

for evento in eventos:
    titulo_tag = evento.find('h3') or evento.find('h2') or evento.find('h1')
    titulo = titulo_tag.get_text(strip=True) if titulo_tag else 'Sin título'

    link_tag = evento.find('a')
    link = link_tag['href'] if link_tag else ''
    if link and not link.startswith('http'):
        link = base_url + link

    img_tag = evento.find('img')
    imagen = img_tag['src'] if img_tag else ''

    # Variables del evento individual
    edad = ubicacion = fecha = categoria = gratuito = 'No disponible'

    if link:
        try:
            detalle = requests.get(link, headers=headers)
            detalle_soup = BeautifulSoup(detalle.content, 'html.parser')
            parrafos = detalle_soup.find_all('p')
            texto_completo = " ".join(p.get_text() for p in parrafos).lower()

            # Edad recomendada
            edad = next((p.get_text() for p in parrafos if 'años' in p.get_text()), 'No disponible')

            # Ubicación
            ubicacion = next((p.get_text() for p in parrafos if 'madrid' in p.get_text().lower() or 'calle' in p.get_text().lower()), 'No disponible')

            # Fechas
            fecha = next((p.get_text() for p in parrafos if 'del' in p.get_text() and 'al' in p.get_text()), 'No disponible')

            # Categoría (si la hubiera en una clase específica)
            categoria_tag = detalle_soup.find('span', class_='posted_in')
            categoria = categoria_tag.get_text(strip=True) if categoria_tag else 'No disponible'

            # Gratuito o no
            if any(palabra in texto_completo for palabra in ['gratuito', 'gratis', 'entrada libre']):
                gratuito = 'Sí'
            elif '€' in texto_completo or 'euros' in texto_completo or 'precio' in texto_completo:
                gratuito = 'No'
            else:
                gratuito = 'No especificado'

            time.sleep(1)  # Para evitar bloquear el servidor

        except Exception as e:
            print(f"Error en {link}: {e}")

    data.append({
        'Título': titulo,
        'Link': link,
        'Imagen': f'<img src="{imagen}" width="120">' if imagen else '',
        'Edad Recomendada': edad,
        'Ubicación': ubicacion,
        'Fechas': fecha,
        'Categoría': categoria,
        '¿Gratuito?': gratuito
    })

# Crear DataFrame
df = pd.DataFrame(data)
display(HTML(df.to_html(escape=False)))
```
