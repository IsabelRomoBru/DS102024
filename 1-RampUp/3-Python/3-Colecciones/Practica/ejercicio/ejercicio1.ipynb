````python
def calcular_precio (fruta:str,kilos:float)->float:
    dicc={ "Manzana":1.5,
          "Pera":2.0,
          "Uva":2.5,
          "plátano":1.0}
    if fruta.capitalize() in dicc.keys():
        return dicc[fruta.capitalize()]* kilos
    else: 
        return "fruta no encontrada en el diccionario"
    
print(calcular_precio("manzana", 5))
````

Conteo de vocales en una cadena
Escribe un programa que tome una cadena de entrada y cuente cuántas vocales (a, e, i, o, u) contiene.

````python
def contador_vocales(c:str):
    contador=0
    for letra in c.lower():
        if letra == "a" or letra=="e" or letra=="i"or letra=="o" or letra=="u":
            contador += 1
    return contador
        
contador_vocales("hola, mundo")
````

````python
def factorial(x:int) -> int:
    if x == 1: 
        return 1
    else:
        return x * factorial (x-1)
factorial(4)
````
