```python
def fetch_all_characters():
    url = "https://swapi.dev/api/people/"
    characters = []
    while url:
        response = requests.get(url)
        data = response.json()
        characters.extend(data['results'])
        url = data['next']  # URL de la siguiente página
        
    return characters
```

```python
def fetch_homeworld(url):
    response = requests.get(url).json()
    return response['name']

def fetch_species(urls):
    if not urls:
        return "Human"  # Asumir "Human" si está vacío
    response = requests.get(urls[0]).json()
    return response['name']

def fetch_films(urls):
    titles = []
    for url in urls:
        response = requests.get(url).json()
        titles.append(response['title'])
    return titles
```

```python
def build_dataframe(characters):
    data = []
    for idx, char in enumerate(characters):
        homeworld = fetch_homeworld(char['homeworld'])
        species = fetch_species(char['species'])
        films = fetch_films(char['films'])

        data.append({
            'id': idx + 1,
            'name': char['name'],
            'height': char['height'],
            'mass': char['mass'],
            'hair_color': char['hair_color'],
            'skin_color': char['skin_color'],
            'eye_color': char['eye_color'],
            'birth_year': char['birth_year'],
            'gender': char['gender'],
            'homeworld': homeworld,
            'species': species,
            'films': films,
        })
    return pd.DataFrame(data)
```

```python
characters = fetch_all_characters()
df = build_dataframe(characters)
df.head()
```
