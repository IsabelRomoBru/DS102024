```python
import numpy as np
import pandas as pd
```

### Paso 2. Importar el conjunto de datos de esta [dirección](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Paso 3. Asígnalo a una variable llamada usuarios.

```python
url='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
```

```python
usuarios=pd.read_csv(url,sep='|')
usuarios.head()
```

### Paso 4. Descubre cuál es la edad media por ocupación

```python
usuarios[['occupation','age']].groupby('occupation').mean().sort_values('age',ascending=False)
```

```python
edadmedia=usuarios[['occupation','age']]
edadmedia.groupby('occupation').mean().sort_values('age',ascending=False)
```

### Paso 5. Descubra la proporción de Hombres por ocupación y ordénela de mayor a menor

#### Forma 1

```python
if 'M' in usuarios['gender'].unique():
    for ocupacion in usuarios['occupation'].sort_values(ascending=False):
        print(ocupacion)
```

#### Forma 2

```python
usuarios.groupby(['occupation','gender'],as_index=False).count()
```

```python
usuarios[usuarios['gender'] == 'M'].sort_values('occupation',ascending=False)
```

```python
hombres= usuarios[usuarios['gender'] == 'M']
hombres.sort_values('occupation',ascending=False)
```

#### Forma 3

```python
hombres = usuarios[usuarios['gender'] == 'M']  
hombres['occupation'].value_counts(normalize=True)  
```

```python
usuarios.groupby(['occupation'])['gender'].value_counts(normalize=True).xs('M',level='gender')
```

### Paso 6. Para cada ocupación, calcule las edades mínima y máxima

```python
usuarios.groupby('age')[['occupation']].aggregate('max','min')
```

```python
usuarios.groupby('age')[['occupation']].max() 
```

### Paso 7. Para cada combinación de ocupación y sexo, calcule la edad media

```python
usuarios.pivot_table(index='occupation',columns='gender',
                     aggfunc={'age':'mean'}, observed=False)
```

### Paso 8.  Para cada ocupación, presente el porcentaje de mujeres y hombres

```python
usuarios.groupby('occupation')['gender'].value_counts(normalize=True)*100
```
