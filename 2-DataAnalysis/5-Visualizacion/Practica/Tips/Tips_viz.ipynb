### Paso 3. Asignarlo a una variable llamada tips

```python
tips=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv')
```

### Paso 4. Trazar el histograma de la columna ``total_bill``

```python
plt.hist(tips['total_bill'], density=True, alpha=0.5);
```

### Paso 5. Cree un gráfico de dispersión que presente la relación entre ``total_bill`` y ``tip``

```python
plt.scatter(x=tips['total_bill'], y=tips['tip']);
plt.title('Relación entre total_bill y tip');
plt.xlabel('Total Bill');
plt.ylabel('Tip');
```

### Paso 6.  Crear una imagen con la relación de `total_bill`, `tip` y `size`.
#### Sugerencia: es sólo una función.

```python
sns.pairplot(tips,vars=['tip','size'],hue='total_bill');
```

### Paso 7. Presente la relación entre ``days`` y ``total_bill``

```python
sns.boxplot(data=tips,x='day', y='total_bill')
plt.title('Relación de bill y los días de la semana')
plt.show();
```

### Paso 8. Crear un gráfico de dispersión con el ``day`` como eje y y ``tips`` como eje x, diferenciar los puntos por ``sex``.

```python
sns.scatterplot(data=tips, x='tip', y='day', hue='sex',palette='pastel')

plt.show();
```

### Paso 9.  Cree una box plot presentación del ``total_bill`` por ``day`` diferetiation el tiempo ``(Dinner or Lunch)``

```python
tips.head()
```

```python
sns.boxplot(data=tips,x='day', y='total_bill',hue='time')
plt.title('Relación de bill y los días de la semana y momento de la comida')
plt.show();
```

### Paso 10. Cree dos histogramas del valor del tip basados en Dinner y Lunch. Deben estar uno al lado del otro.

```python
ld=sns.FacetGrid(data=tips, col="time", height=4, aspect=1.2)
ld.map(sns.histplot,"tip",kde=False,color="skyblue")
ld.set_axis_labels("Tip ", "Frecuencia")
ld.set_titles("{col_name}")  
ld.tight_layout()

plt.show()
```

### Paso 11. Cree dos gráficos de dispersión, uno para los hombres y otro para las mujeres, que presenten la relación entre el valor ``total_bill`` y la ``tip``, diferenciando entre ``smoker`` y ``no smoker``.
### Deben estar uno al lado del otro.

```python
g = sns.FacetGrid(data=tips, col="sex", height=4, aspect=1.2)
g.map(sns.scatterplot,data=tips, x="total_bill", y="tip", hue="smoker", palette="coolwarm")

# etiquetas y títulos
g.set_axis_labels("Total Bill", "Tip Amount")
g.add_legend(title="Smoker Status")
g.set_titles("{col_name}")

# Mostrar gráfico
g.tight_layout()
plt.show()
```
