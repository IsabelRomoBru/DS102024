```python
import numpy as np
import pandas as pd
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv')
titanic
```

```python
df=titanic[['Pclass','Name','Sex','Age','SibSp','Parch','Fare','Embarked','Survived']]
```

Imputaciones + Feature Engineer+

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

Feature engineer

```python
#df['Title']=df['Name'].apply(lambda x: x.split(', ')[1].split('.')[0])
#df['Title']=df['Name'].apply(lambda x: x.split('.')[0]) ',(\w+)\.'

df['Title']=df['Name'].str.extract(', (\w+)\.').fillna('Sin titulo')
df.drop(columns='Name',inplace=True)
```
