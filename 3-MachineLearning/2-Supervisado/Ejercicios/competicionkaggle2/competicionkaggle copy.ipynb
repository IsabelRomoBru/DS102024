## Transformación de datos

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Eliminar columnas irrelevantes
#df_cleaned = df.drop(columns=["Unnamed: 0", "id"])

# Manejar valores nulos en variables categóricas reemplazando con "Desconocido"
#df_cleaned = df_cleaned.fillna("Desconocido")

# Aplicar Label Encoding a todas las columnas categóricas
label_encoders = {}
for col in df.columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Guardar el encoder para referencia

# Ver las primeras filas después de la transformación
df.head()
```

## K-Means

```python
# Normalizar los datos para K-Means
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans
# Probar diferentes valores de K usando el método del codo
costs = []
K_range = range(2, 11)

for K in K_range:
    kmeans = KMeans(n_clusters=K, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    costs.append(kmeans.inertia_)  # Inercia (costo)
# Graficar el método del codo
plt.figure(figsize=(8, 5))
plt.plot(K_range, costs, marker="o", linestyle="--", color="b")
plt.xlabel("Número de Clusters (K)")
plt.ylabel("Costo (Inercia)")
plt.title("Método del Codo para K-Means")
plt.xticks(K_range)
plt.grid(True)
plt.show();
```

```python
# Aplicar K-Means con K=7
kmeans = KMeans(n_clusters=8, random_state=42, n_init=10)
df["cluster"] = kmeans.fit_predict(df_scaled)

# Ver distribución de los clusters
cluster_counts = df["cluster"].value_counts()
cluster_counts
```

```python
print("Inercia (costo total):", kmeans.inertia_)
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(df, df["cluster"])
print("Silhouette Score:", sil_score)
```
