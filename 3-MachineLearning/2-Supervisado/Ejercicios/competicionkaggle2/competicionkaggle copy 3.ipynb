## Transformación de datos

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
# Identificar columnas categóricas
categorical_cols = df.select_dtypes(include=['object']).columns
    
# Aplicar LabelEncoder a las columnas categóricas
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
df
    
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import GradientBoostingClassifier

# Separamos características y target
X_train = df.drop('class', axis=1)
y_train = df['class']


# 2. Entrenamiento del modelo Gradient Boosting Classifier
# Entrenar modelo de Gradient Boosting
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbc.fit(X_train, y_train)

```

```python
test.columns = test.columns.str.replace("-", "_")
test
```

```python
# Asegurar que los valores nulos se llenen correctamente sin advertencias
df_test = test.copy()  # Evitar la advertencia sobre copias en Pandas

for col in test.columns:
    df_test[col] = df_test[col].fillna(df_test[col].mode()[0])  # Asignación correcta

```

```python
# Volver a aplicar Label Encoding con manejo de valores no vistos
for col in df_test.columns:
    if col in label_encoders:
        le = label_encoders[col]
        df_test[col] = df_test[col].apply(lambda x: le.transform([x])[0] if x in le.classes_ else -1)
```

```python
from sklearn.metrics import f1_score
# Asegurar que las columnas de test coincidan en orden y nombres con las de entrenamiento
df_test = df_test[X_train.columns]  # Reordenar y seleccionar solo las columnas usadas en entrenamiento

# Generar predicciones con el modelo entrenado
test_predictions = gbc.predict(df_test)

```

```python
from sklearn.model_selection import cross_val_score

# Opción 2: Cross-validation en todo `train.csv` (5 folds)
f1_cv = cross_val_score(gbc, X_train, y_train, cv=5, scoring="f1_weighted").mean()

# Mostrar resultados
f1_cv
```
