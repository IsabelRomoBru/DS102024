## Transformación de datos

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
def encode_and_scale(train_data):
    # Identificar columnas categóricas
    categorical_cols = train_data.select_dtypes(include=['object']).columns
    
    # Aplicar LabelEncoder a las columnas categóricas
    label_encoders = {}
    for col in categorical_cols:
        le = LabelEncoder()
        train_data[col] = le.fit_transform(train_data[col])
        label_encoders[col] = le
    
    # Normalizar los datos
    scaler = StandardScaler()
    train_scaled = scaler.fit_transform(train_data)
    
    print("Codificación y escalado completados.")
    return train_scaled, label_encoders, scaler
```

```python
def apply_pca(train_scaled, n_components=2):
    # Aplicar PCA
    pca = PCA(n_components=n_components)
    train_pca = pca.fit_transform(train_scaled)
    
    # Convertir a DataFrame
    train_pca_df = pd.DataFrame(train_pca, columns=[f'PC{i+1}' for i in range(n_components)])
    
    # Guardar los datos transformados
    train_pca_df.to_csv('train_pca.csv', index=False)
    
    print("PCA aplicado. Datos guardados como 'train_pca.csv'")
    return train_pca_df, pca
```

```python

# Ejecutar las funciones
train_scaled, encoders, scaler = encode_and_scale(train_data)
train_pca, pca = apply_pca(train_scaled)
```
