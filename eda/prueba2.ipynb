descripción
Este conjunto de datos contiene información sobre la violencia doméstica contra las mujeres en una zona rural específica de un país en desarrollo. Los datos se recopilan para ayudar a comprender la correlación entre varios factores socioeconómicos y la violencia doméstica.

Contenido del conjunto de datos
El conjunto de datos incluye las siguientes columnas:

SL. No: Número de serie del registro.
Edad: Edad del encuestado.
Educación: Nivel educativo del encuestado (terciario para secundario superior).
Empleo: Situación laboral del demandado.
Ingresos: Nivel de ingresos del encuestado (0 que indica que no hay ingresos).
Estado civil: Estado civil del demandado (casado o no casado).
Violencia: Indica si el encuestado ha sufrido violencia doméstica (sí o no).
Contexto
La violencia doméstica contra las mujeres es un problema importante en muchos países en desarrollo. Comprender los factores que contribuyen a esta violencia puede ayudar a crear intervenciones y políticas efectivas. Este conjunto de datos tiene como objetivo proporcionar una base para el análisis y la investigación en esta área crítica.
Inspiración
El conjunto de datos se puede utilizar para:

Analizar la correlación entre el nivel de educación y la violencia doméstica.
Estudie el impacto del estado civil en la violencia doméstica.
Explore la distribución por edades de las víctimas de violencia doméstica.
Identificar los posibles factores socioeconómicos que contribuyen a la violencia doméstica.
