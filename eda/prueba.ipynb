## Características del conjunto de datos

1. País: Nombre del país.
2. Código: Código de país ISO.
3. Región: Región geográfica.
4. régimen_político: Tipo de régimen político.
5. safety_and_security: Puntuación por el grado en que la guerra, el conflicto, el terror y el crimen han desestabilizado la seguridad.
6. libertad_personal: Puntuación para el progreso hacia los derechos legales, las libertades individuales y la tolerancia social.
7. gobernanza: Puntuación para el alcance de los controles de poder, la eficacia del gobierno y la falta de corrupción.
8. capital_social: Puntuación de la fuerza de las relaciones personales/sociales, confianza institucional, normas sociales y participación cívica.
9. investment_environment: Puntuación para la protección y accesibilidad de las inversiones.
10. enterprise_conditions: Puntuación para las regulaciones que permiten el inicio del negocio, la competencia y la expansión.
11. infrastructure_and_market_access: Puntuación de la calidad de la infraestructura comercial y las condiciones del mercado.
12. calidad_económica: Puntuación para la generación sostenible de riqueza y el compromiso de la fuerza laboral.
13. condiciones de vida: Puntuación de la calidad de vida, incluidos los recursos materiales, el refugio, los servicios básicos y la conectividad.
14. salud: Puntuación para resultados de salud, sistemas, factores de enfermedad/riesgo y tasas de mortalidad.
15. Educación: Puntuación para la inscripción, los resultados y la calidad en todas las etapas de la educación y las habilidades de los adultos.
16. natural_ambiente: Puntuación de los aspectos del entorno físico que afectan la vida diaria y la prosperidad futura.
17. average_score: Puntuación media general.
